```python
file_path = "menu.csv"  # Update this path to the location of your file
output_file_path = "cleaned_menu.csv"

# Read the CSV file
menu_data = pd.read_csv(file_path)

# Step 1: Extract numeric values from the 'price' column
menu_data["price_cleaned"] = (
    menu_data["price"].astype(str).str.extract(r"(\d+\.?\d*)")[0]
)

# Step 2: Convert 'price_cleaned' to float
menu_data["price_cleaned"] = menu_data["price_cleaned"].astype(float)

# Step 3: Handle missing or problematic values (if needed)
# Replace NaN values with 0 or another default value (optional)
menu_data["price_cleaned"].fillna(0, inplace=True)

# Step 4: Save the cleaned data to a new CSV file
menu_data.to_csv(output_file_path, index=False)

print(f"Cleaned data has been saved to {output_file_path}")
```

```python
users_df = pd.read_csv("users.csv")


# Define mapping function
def categorize_income(income):
    if income == "No Income":
        return "No Income"
    elif "Below Rs.10000" in income:
        return "Below 10K"
    elif "10001 to 25000" in income:
        return "10K to 25K"
    elif "25001 to 50000" in income:
        return "25K to 50K"
    elif "More than 50000" in income:
        return "More than 50K"
    else:
        return "Unknown"


# Apply categorization
users_df["Income Bracket"] = users_df["Monthly Income"].apply(categorize_income)

# Display results
print(users_df.head())
```
